1\. Plot the functions:
- $f_1(x)=\frac{1}{1+e^{-0.4x}}$, in blue;
- $f_2(x)=\frac{1}{1+e^{-0.7x}}$, in red;
- $f_3(x)=\frac{1}{1+e^{-x}}$, in green;
- $f_4(x)=\frac{1}{1+e^{-2x}}$, in black.

from -6 to 6, in a figure 7x4 inches big. 

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-6,6,100)
y1 = 1 / (1 + np.exp(-1*0.4*x))
y2 = 1 / (1 + np.exp(-1*0.7*x))
y3 = 1 / (1 + np.exp(-1*x))
y4 = 1 / (1 + np.exp(-2*x))

fig, axes = plt.subplots(figsize=(7,4))

axes.plot(x, y1, 'b', label = "$f_1(x)$")
axes.plot(x, y2, 'r', label = '$f_2(x)$')
axes.plot(x, y3, 'g', label = '$f_3(x)$')
axes.plot(x, y4, 'black', label = '$f_4(x)$')
```

2\. Make the y-axis limits range from 0 to 1, and the x-axis limits to be as tight as possible.

```python
axes.axis('tight')
axes.set_ylim([0,1])
fig
```

3\. Do the following:

- Add a legend that indicates which line is which;
- Add the x-label "Input";
- Add the y-label "Output";
- Add a dashed grid;
- Add the title "Activations".

```python
# Add legend
axes.legend()

# Add x and y labels
axes.set_xlabel("Input")
axes.set_ylabel("Output")

# Add dashed grid
axes.grid(True, linestyle='--', alpha=0.7)

# Add title
axes.set_title("Activations")
fig
```

4\. Change the y ticks to be only 0, 0.5 and 1. Change the x ticks to be -6, -3, 0, 3, and 6.

```python
axes.set_xticks([-6, -3, 0, 3, 6])
axes.set_yticks([0, 0.5, 1])
fig
```

5\. Create a new figure, this time with a 2x2 axes layout. Create the same plot as you did so far, but each line should now be separated into one axis.

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

x = np.linspace(-6,6,100)
y1 = 1 / (1 + np.exp(-1*0.4*x))
y2 = 1 / (1 + np.exp(-1*0.7*x))
y3 = 1 / (1 + np.exp(-1*x))
y4 = 1 / (1 + np.exp(-2*x))


axes[0][0].plot(x, y1, 'b', label = "$f_1(x)$")
axes[0][1].plot(x, y2, 'r', label = '$f_2(x)$')
axes[1][0].plot(x, y3, 'g', label = '$f_3(x)$')
axes[1][1].plot(x, y4, 'black', label = '$f_4(x)$')

# Add legend
axes[0][0].legend()
axes[0][1].legend()
axes[1][0].legend()
axes[1][1].legend()
```
