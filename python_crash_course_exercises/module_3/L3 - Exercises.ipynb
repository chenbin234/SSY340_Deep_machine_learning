1\. Plot a histogram with 30 bins of 300k samples from a Beta distribution with parameters $\alpha=2$ and $\beta=5$. Give it a title and label the axes.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate samples from a Beta distribution
alpha = 2
beta = 5
num_samples = 300000
samples = np.random.beta(alpha, beta, num_samples)

# Create the histogram
plt.hist(samples, bins=30, alpha=0.7, color='blue')

# Add title and labels
plt.title("Histogram of Beta Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

2\. Generate the following plot: ![](pretty_plot.png)

```python
fig, ax = plt.subplots(figsize=(8,4))

x = np.linspace(-np.pi, np.pi, 200)
y1 = np.sin(x)
y2 = np.sin(x-np.pi/4)

ax.plot(x, y1, 'r', label='$sin(x)$')
ax.plot(x, y2, 'black', label='$sin(x-\\frac{\\pi}{4})$')
ax.fill_between(x, y1, y2)

ax.set_xlim([-np.pi, np.pi])
ax.set_ylim([-1,1])
ax.legend(loc=4)
ax.set_title('Difference in phase angle')
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_yticks([-1, 0, 1])

ax.grid(linestyle='dashed', linewidth=0.5)


ax.set_yticklabels([-1, 0, 1])
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', 0, r'$\pi/2$', r'$\pi$'])


ax.set_xlabel("x")
ax.set_ylabel("y")
```

3\. Generate a contour plot for the function $$f(x,y)=\frac{\sin(5x)\cos(5y)}{5}.$$

- x range: $-\pi/5$ to $\pi/5$;
- y range: $-\pi/7$ to $\pi/7$;
- 20 different levels.

```python
# Define the function
def f(x, y):
    return (np.sin(5*x) * np.cos(5*y)) / 5

# Create a grid of x and y values
x = np.linspace(-np.pi/5, np.pi/5, 400)
y = np.linspace(-np.pi/7, np.pi/7, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create the contour plot
plt.contour(X, Y, Z, levels=20, cmap='viridis')

# Add x and y labels
plt.xlabel("x")
plt.ylabel("y")

# Add a colorbar
plt.colorbar()

# Add title
plt.title("Contour Plot of f(x, y)")

# Display the plot
plt.show()
```
